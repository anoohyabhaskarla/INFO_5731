The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
# Your code here
import math

def calculate_hemisphere_volume(radius):
    volume = (2/3) * math.pi * radius**3
    return volume
print("r    V")

for r in range(2, 21, 2):
    volume = calculate_hemisphere_volume(r)
    print(f"{r}    {volume}")

```

(b) Arrange the generated numbers in ascending order. Display the sorted list.

(d) Compute and output the average of the entire group of generated numbers.

(e) Enhance the program to calculate and display the average of the first 6 numbers from both the ascending and descending sorted lists.

```python
# Write code for each of the sub parts with proper comments.
import random

def generate_random_sequence(length, minimum, maximum):
    return [random.randint(minimum, maximum) for _ in range(length)]

def calculate_average(numbers):
    return sum(numbers) / len(numbers)

# Set the parameters
sequence_length = 12
minimum_value = 50
maximum_value = 150

# (a) Generate the random sequence
random_sequence = generate_random_sequence(sequence_length, minimum_value, maximum_value)

# (b) Sort the sequence in ascending order
ascending_sorted_sequence = sorted(random_sequence)
print("Sorted Random Sequence (Ascending):", ascending_sorted_sequence)

# (c) Sort the sequence in descending order
descending_sorted_sequence = sorted(random_sequence, reverse=True)
print("Sorted Random Sequence (Descending):", descending_sorted_sequence)

# (d) Compute and output the average of the entire group of generated numbers
average_whole_group = calculate_average(random_sequence)
print("Average of the entire group:", average_whole_group)

# (e) Compute and display the average of the first 6 numbers from both ascending and descending sorted lists
average_first_six_ascending = calculate_average(ascending_sorted_sequence[:6])
average_first_six_descending = calculate_average(descending_sorted_sequence[:6])

print("Average of the first 6 numbers (Ascending):", average_first_six_ascending)
print("Average of the first 6 numbers (Descending):", average_first_six_descending)

```

In the field of Data Science, data is often formatted as a comma-delimited (CSV) file, where each line in a file corresponds to a field's value. Refer to Chapter 9 (p. 106) for more information on comma-delimited files.

The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

(a) (5 pts) Calculate the difference in the number of Occupied Housing Units from year to year and print it. The difference must be calculated for the consecutive years such as 2008-2009, 2009-2010 etc. Finally, print the values in the ascending order.

(b) (5 pts) For all the years, calculate the percentage of housing units which are vacant an occupied. Print the results in the following format:

Year Vacant Housing Units Occupied Housing Units

    2008  30%   70%                          
	2009 ----- -----
	2010 ----- -----
    2011 ----- -----
	2012 ----- -----
    2013 ----- -----
	2014 ----- -----
  
  

(c) (5 pts) Calculate and print the valued and years in which the highest number of housing units were vacant and occupied. Print the results in the following format:  

                              Year	Value
	Vacant Housing Units	  -----   -----
	Occupied Housing Units	-----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
import statistics

# a) Function to compute year-over-year differences in occupied housing units
def compute_yearly_differences(data):
    occupied_data = data[data['title_field'] == 'Occupied Housing Units']
    occupied_data = occupied_data[['year', 'value']].sort_values(by='year')
    differences = occupied_data['value'].diff().dropna()
    sorted_differences = differences.sort_values()

    print("\nSorted Yearly Differences in Occupied Housing Units:")
    for i, diff in enumerate(sorted_differences, start=1):
        print(f"{occupied_data.iloc[i-1]['year']}-{occupied_data.iloc[i]['year']}: {diff}")
    return sorted_differences

# b) Function to calculate and display the percentage of vacant vs. occupied housing units
def calculate_unit_percentages(data):
    years = data['year'].unique()
    print(f"\n{'Year':<10} {'Vacant Units (%)':<20} {'Occupied Units (%)':<25}")
    for year in years:
        total_units = data[data['year'] == year]['value'].sum()
        vacant_units = data[(data['year'] == year) & (data['title_field'] == 'Vacant Housing Units')]['value'].values[0]
        occupied_units = data[(data['year'] == year) & (data['title_field'] == 'Occupied Housing Units')]['value'].values[0]
        vacant_pct = (vacant_units / total_units) * 100
        occupied_pct = (occupied_units / total_units) * 100
        print(f"{year:<10} {vacant_pct:<20.2f} {occupied_pct:<25.2f}")

# c) Function to find and print the year with maximum vacant and occupied units
def max_vacant_occupied_year(data):
    vacant_data = data[data['title_field'] == 'Vacant Housing Units']
    occupied_data = data[data['title_field'] == 'Occupied Housing Units']
    max_vacant_year = vacant_data.loc[vacant_data['value'].idxmax()]
    max_occupied_year = occupied_data.loc[occupied_data['value'].idxmax()]

    print("\nYear with Maximum Vacant and Occupied Units:")
    print(f"{'Category':<25} {'Year':<10} {'Units':<10}")
    print(f"Vacant Units               {max_vacant_year['year']:<10} {max_vacant_year['value']:<10}")
    print(f"Occupied Units             {max_occupied_year['year']:<10} {max_occupied_year['value']:<10}")

# d) Function to compute the harmonic mean of total housing units
def harmonic_mean_total_units(data):
    yearly_totals = data.groupby('year')['value'].sum()
    harmonic_mean = statistics.harmonic_mean(yearly_totals)
    print(f"\nHarmonic Mean of Total Housing Units: {harmonic_mean:.2f}")

# Main function to load data and call the relevant analysis functions
def main():
    # Loading data from the provided CSV file
    data = pd.read_csv('Assignment1_denton_housing.csv')

    # (a) Compute and print year-over-year differences in occupied housing units
    compute_yearly_differences(data)

    # (b) Calculate and display the percentage of vacant and occupied housing units by year
    calculate_unit_percentages(data)

    # (c) Find the year with the highest number of vacant and occupied housing units
    max_vacant_occupied_year(data)

    # (d) Compute the harmonic mean of total housing units over the years
    harmonic_mean_total_units(data)

if __name__ == "__main__":
    main()

```

Prompt 3: Improve the inventory system by incorporating search capabilities. Give customers the option to look for products by Product ID, Name, or Category. To increase usability, use partial matches and case-insensitive searches. Make sure the system responds meaningfully in the event that a product cannot be located.

```python
import csv

class InventoryManager:
    def __init__(self, filename="inventory.csv"):
        self.filename = filename
        self.inventory = []
        self.load_inventory()

    def load_inventory(self):
        """Load inventory data from a CSV file."""
        try:
            with open(self.filename, mode='r', newline='') as file:
                reader = csv.DictReader(file)
                self.inventory = list(reader)
        except FileNotFoundError:
            self.inventory = []
        except Exception as e:
            print(f"Error loading inventory: {e}")

    def save_inventory(self):
        """Save inventory data to a CSV file."""
        with open(self.filename, mode='w', newline='') as file:
            fieldnames = ["Product ID", "Name", "Category", "Quantity", "Price", "Supplier"]
            writer = csv.DictWriter(file, fieldnames=fieldnames)
            writer.writeheader()
            writer.writerows(self.inventory)

    def add_product(self, product_id, name, category, quantity, price, supplier):
        """Add a new product to the inventory."""
        self.inventory.append({
            "Product ID": product_id,
            "Name": name,
            "Category": category,
            "Quantity": quantity,
            "Price": price,
            "Supplier": supplier
        })
        self.save_inventory()

    def delete_product(self, product_id):
        """Delete a product by Product ID."""
        self.inventory = [p for p in self.inventory if p["Product ID"] != product_id]
        self.save_inventory()

    def update_product(self, product_id, field, new_value):
        """Update a product's field with a new value."""
        for product in self.inventory:
            if product["Product ID"] == product_id:
                product[field] = new_value
                break
        self.save_inventory()

    def search_product(self, search_term):
        """Search for products by Product ID, Name, or Category (case-insensitive, partial match)."""
        search_term = search_term.lower()
        results = [p for p in self.inventory if search_term in p["Product ID"].lower() or
                   search_term in p["Name"].lower() or
                   search_term in p["Category"].lower()]

        if results:
            for product in results:
                print(product)
        else:
            print("No products found matching the search criteria.")

# Example Usage
if __name__ == "__main__":
    inventory = InventoryManager()
    inventory.add_product("001", "Laptop", "Electronics", "10", "1200", "TechCorp")
    inventory.add_product("002", "Smartphone", "Electronics", "25", "800", "MobileMakers")
    inventory.search_product("laptop")  # Should return the laptop entry
    inventory.search_product("Electronics")  # Should return all electronics
```

```python
import random
import csv
import os

# Constants
CSV_FILE = "flights.csv"
SEAT_CAPACITY = 150  # Default seat capacity for each flight

# Generate 50 flights with random data
def generate_flights():
    flights = []
    cities = ["New York", "Los Angeles", "Chicago", "Houston", "Miami", "Seattle", "San Francisco", "Denver", "Boston", "Las Vegas"]
    for i in range(50):
        flight_id = f"F{random.randint(1000, 9999)}"
        departure = random.choice(cities)
        destination = random.choice([c for c in cities if c != departure])
        seats_available = random.randint(50, SEAT_CAPACITY)
        price = round(random.uniform(100, 1000), 2)
        flights.append([flight_id, departure, destination, seats_available, price])
    return flights

# Initialize CSV file with flights if not exist
def initialize_flights():
    if not os.path.exists(CSV_FILE):
        flights = generate_flights()
        with open(CSV_FILE, "w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Departure", "Destination", "SeatsAvailable", "Price"])
            writer.writerows(flights)
        print("Flight data initialized.")

# Load flights from CSV
def load_flights():
    with open(CSV_FILE, "r") as file:
        reader = csv.DictReader(file)
        return list(reader)

# Save flights back to CSV
def save_flights(flights):
    with open(CSV_FILE, "w", newline="") as file:
        writer = csv.writer(file)
        writer.writerow(["FlightID", "Departure", "Destination", "SeatsAvailable", "Price"])
        for flight in flights:
            writer.writerow([flight["FlightID"], flight["Departure"], flight["Destination"], flight["SeatsAvailable"], flight["Price"]])

# Search for available flights
def search_flights():
    departure = input("Enter departure city: ").strip()
    destination = input("Enter destination city: ").strip()
    flights = load_flights()

    results = [f for f in flights if f["Departure"].lower() == departure.lower() and f["Destination"].lower() == destination.lower()]

    if results:
        print("Available Flights:")
        for flight in results:
            print(f"Flight {flight['FlightID']} - Seats: {flight['SeatsAvailable']}, Price: ${flight['Price']}")
    else:
        print("No flights available for the selected route.")

# Book a seat on a flight
def book_flight():
    flight_id = input("Enter Flight ID to book: ").strip()
    flights = load_flights()

    for flight in flights:
        if flight["FlightID"] == flight_id:
            if int(flight["SeatsAvailable"]) > 0:
                flight["SeatsAvailable"] = str(int(flight["SeatsAvailable"]) - 1)
                save_flights(flights)
                print(f"Booking successful! Seat reserved on Flight {flight_id}.")
            else:
                print("Flight is fully booked.")
            return
    print("Flight not found.")

# Cancel a booking
def cancel_booking():
    flight_id = input("Enter Flight ID to cancel booking: ").strip()
    flights = load_flights()

    for flight in flights:
        if flight["FlightID"] == flight_id:
            flight["SeatsAvailable"] = str(int(flight["SeatsAvailable"]) + 1)
            save_flights(flights)
            print(f"Booking canceled. Seat released for Flight {flight_id}.")
            return
    print("Flight not found.")

# Generate a report of flight availability
def generate_report():
    flights = load_flights()
    print("\nFlight Report:")
    for flight in flights:
        print(f"{flight['FlightID']} - {flight['Departure']} to {flight['Destination']} - Available Seats: {flight['SeatsAvailable']}")

# Main Menu
def main():
    initialize_flights()
    while True:
        print("\nFlight Booking System")
        print("1. Search Flights")
        print("2. Book a Flight")
        print("3. Cancel a Booking")
        print("4. Generate Report")
        print("5. Exit")
        choice = input("Select an option: ").strip()

        if choice == "1":
            search_flights()
        elif choice == "2":
            book_flight()
        elif choice == "3":
            cancel_booking()
        elif choice == "4":
            generate_report()
        elif choice == "5":
            print("Exiting system. Goodbye!")
            break
        else:
            print("Invalid option. Please try again.")

if __name__ == "__main__":
    main()
''
```

```python
# Your code here
import csv
import os
import random

# Filepath for flights CSV file
FLIGHTS_FILE = "flights.csv"

# Initialize the flight system
def initialize_flights():
    if not os.path.exists(FLIGHTS_FILE):
        with open(FLIGHTS_FILE, mode='w', newline='') as file:
            writer = csv.writer(file)
            writer.writerow(["Flight ID", "Airline", "Origin", "Destination", "Seats Available", "Total Seats", "Price"])
        populate_initial_flights()

# Load flights from CSV
def load_flights():
    flights = []
    try:
        with open(FLIGHTS_FILE, mode='r') as file:
            reader = csv.DictReader(file)
            for row in reader:
                flights.append(row)
    except Exception:
        print("Error reading flight file. Initializing a new flight system.")
        initialize_flights()
    return flights

# Add a new flight
def add_flight(flight_id, airline, origin, destination, seats_available, total_seats, price):
    flights = load_flights()
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            print("Error: Flight ID already exists.")
            return
    with open(FLIGHTS_FILE, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerow([flight_id, airline, origin, destination, seats_available, total_seats, price])
    print("Flight added successfully.")

# Find flights by origin and destination
def find_flight(origin, destination):
    flights = load_flights()
    results = [flight for flight in flights if flight["Origin"].lower() == origin.lower() and flight["Destination"].lower() == destination.lower()]
    return results

# Reserve a seat
def reserve_seat(flight_id):
    flights = load_flights()
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Seats Available"]) > 0:
                flight["Seats Available"] = str(int(flight["Seats Available"]) - 1)
                save_flights(flights)
                print("Seat reserved successfully.")
                return
            else:
                print("Error: No available seats.")
                return
    print("Error: Flight ID not found.")

# Release a seat (Cancellation)
def release_seat(flight_id):
    flights = load_flights()
    for flight in flights:
        if flight["Flight ID"] == flight_id:
            if int(flight["Seats Available"]) < int(flight["Total Seats"]):
                flight["Seats Available"] = str(int(flight["Seats Available"]) + 1)
                save_flights(flights)
                print("Seat cancellation successful.")
                return
            else:
                print("Error: Cannot release more seats than total available.")
                return
    print("Error: Flight ID not found.")

# Generate performance reports
def generate_report():
    flights = load_flights()
    total_revenue = sum(int(flight["Total Seats"]) * float(flight["Price"]) for flight in flights)
    print("Total Revenue from Flights: $", round(total_revenue, 2))
    print("Flight Summary:")
    for flight in flights:
        print(f"{flight['Airline']} | {flight['Origin']} -> {flight['Destination']} | Seats Left: {flight['Seats Available']}")

# Save updated flight records
def save_flights(flights):
    with open(FLIGHTS_FILE, mode='w', newline='') as file:
        writer = csv.DictWriter(file, fieldnames=["Flight ID", "Airline", "Origin", "Destination", "Seats Available", "Total Seats", "Price"])
        writer.writeheader()
        writer.writerows(flights)

# Populate initial flights with 50 different records
def populate_initial_flights():
    airlines = ["Delta", "United", "American Airlines", "Southwest", "JetBlue"]
    cities = ["New York", "Los Angeles", "Chicago", "Houston", "Miami", "San Francisco"]
    sample_flights = [
        (f"F{100+i}", random.choice(airlines), random.choice(cities), random.choice([c for c in cities if c != cities[i % len(cities)]]), random.randint(10, 200), 200, round(random.uniform(100, 1000), 2))
        for i in range(50)
    ]
    with open(FLIGHTS_FILE, mode='a', newline='') as file:
        writer = csv.writer(file)
        writer.writerows(sample_flights)

# Ensure flight system maintains at least 50 flights
def maintain_flights():
    flights = load_flights()
    if len(flights) < 50:
        populate_initial_flights()

# Initialize flight system and maintain 50 flights
initialize_flights()
maintain_flights()

# Test functions
add_flight("F150", "Alaska Airlines", "Boston", "Seattle", 50, 150, 350.00)
reserve_seat("F100")
release_seat("F100")
print("Flight Search Results:", find_flight("New York", "Los Angeles"))
generate_report()

```
